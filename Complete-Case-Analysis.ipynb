```python
# the following function captures the percentage of observations
# for each category in the original and complete case dataset
# and puts them together in a new dataframe
class Graph_c():
    def __init__(self,df1,df2,var):
        self.df1=df1
        self.df2=df2
        self.var=var
    ## let's check the distribution of a few variables before and after, cca: histogram
    def grph_hist(self):
        fig=plt.figure()
        ax=fig.add_subplot(111)
        return self.df1[self.var].hist(bins=50, ax=ax, density=True, color='red'), self.df2[self.var].hist(bins=50, ax=ax, density=True, color='blue')
    ## let's check the distribution of a few variables before and after 
    # cca: density plot
    def graph_den(self):
        fig = plt.figure()
        ax = fig.add_subplot(111)
        return self.df1[self.var].plot.density(color='red'), self.df2[self.var].plot.density(color='blue')
        
    def cat_dist(self,df1,df2,var):
        C=pd.concat([df1[var].value_counts()/len(df1[var]),df2[var].value_counts()/len(df2[var])],axis=1)
        C.columns=['original', 'cca']
        C['diff']=C['cca']-C['original']
        return C
```
